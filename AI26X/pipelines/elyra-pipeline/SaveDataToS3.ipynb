```python
import boto3
import os
import pandas as pd
```

```python
s3_endpoint_url = os.environ.get("AWS_S3_ENDPOINT")
s3_access_key = os.environ.get("AWS_ACCESS_KEY_ID")
s3_secret_key = os.environ.get("AWS_SECRET_ACCESS_KEY")
s3_bucket_name = os.environ.get("AWS_S3_BUCKET")

print(
    f"Uploading data"
    f'from bucket "{s3_bucket_name}" '
    f"from S3 storage at {s3_endpoint_url}"
)

s3_client = boto3.client(
    "s3",
    endpoint_url=s3_endpoint_url,
    aws_access_key_id=s3_access_key,
    aws_secret_access_key=s3_secret_key,
)
```

```python
def upload_df_s3(local_folder, file_name, s3_folder):
    full_local_file_name = local_folder + "/" + file_name
    full_s3_file_name = s3_folder + "/" + file_name

    s3_client.upload_file(
        full_local_file_name, s3_bucket_name, Key=full_s3_file_name)

```

# Read from local storage

```python
local_folder = './resources/data/'
s3_folder = '/fpg_test/'

for file in os.listdir(local_folder):
    if file.endswith('.csv'):
        upload_df_s3(local_folder, file, s3_folder)
```

# Test Upload

```python
print("Commencing data ingestion.")

# List objects within the specified bucket and folder
response = s3_client.list_objects_v2(Bucket=s3_bucket_name, Prefix=s3_folder)

for obj in response.get("Contents", []):
    key = obj["Key"]
    file_name = os.path.basename(key)
    print(file_name)
```
